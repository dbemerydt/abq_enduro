```python
import numpy as np
import pandas as pd
```

```python
raw_results = pd.read_csv('example_input.csv')
raw_results
```

```python
num_stages = 3
df = raw_results.copy()
for i in range(1,num_stages+1):
    df['stage_'+str(i)+'_start'] = pd.to_datetime(df['stage_'+str(i)+'_start'],format='%M:%S')
    df['stage_'+str(i)+'_finish'] = pd.to_datetime(df['stage_'+str(i)+'_finish'],format='%M:%S')
df
```

```python
for i in range(1,num_stages+1):
    df['stage_'+str(i)+'_time'] = df['stage_'+str(i)+'_finish'] - df['stage_'+str(i)+'_start']
    df['stage_'+str(i)+'_time_rank'] = df['stage_'+str(i)+'_time'].rank(ascending=True)
df
```

```python
df['total_time'] = df[['stage_'+str(i)+'_time' for i in range(1,num_stages+1)]].sum(axis=1)
df
```

```python
df['rank'] = df['total_time'].rank(ascending=True)
df
```

```python
for i in range(1,num_stages+1):
    df['stage_'+str(i)+'_start'] = df['stage_'+str(i)+'_start'].astype('str').str.slice(10)
    df['stage_'+str(i)+'_finish'] = df['stage_'+str(i)+'_finish'].astype('str').str.slice(10)
df
```
